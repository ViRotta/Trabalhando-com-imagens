```python
from numpy.core.fromnumeric import shape
import matplotlib.pyplot as plt
from skimage.io import imread,imsave, imshow
import numpy as np


image = imread ('/content/Lena.jpg')

red = image[:,:,0]
green = image[:,:,1]
blue = image[:,:,2]
gray = image [:,:,0]

for i in range(image.shape[0]):
  for j in range(image.shape[1]):

    x = red[i][j] + green[i][j] + blue[i][j]

gray [i][j]= x/3

plt.figure()
plt.imshow(gray,cmap = 'gray')

img = np.zeros((image.shape[0],image.shape[1],3),'uint8')

img[:,:,0] = gray
img[:,:,1] = gray
img[:,:,2] = gray

imsave('/content/cinza.png', img)
```

Reduzindo e restaurando por interpola√ß√£o

```python
from PIL import Image
import matplotlib.pyplot as plt

def reduzir_imagem(imagem, escala):
    largura = int(imagem.width * escala)
    altura = int(imagem.height * escala)
    imagem_redimensionada = imagem.resize((largura, altura), Image.LANCZOS)
    return imagem_redimensionada

# Carregando a imagem original
imagem_original = Image.open("/content/Lena.jpg")

# Reduzindo a imagem em diferentes escalas
escala_50 = 0.5
imagem_reduzida_50 = reduzir_imagem(imagem_original, escala_50)
imsave('/content/imagem_reduzida_50.png', img)


escala_25 = 0.25
imagem_reduzida_25 = reduzir_imagem(imagem_original, escala_25)
imsave('/content/imagem_reduzida_25.png', img)


escala_12_5 = 0.125
imagem_reduzida_12_5 = reduzir_imagem(imagem_original, escala_12_5)
imsave('/content/imagem_reduzida_12_5.png', img)


escala_6_25 = 0.0625
imagem_reduzida_6_25 = reduzir_imagem(imagem_original, escala_6_25)
imsave('/content/imagem_reduzida_6_25.png', img)


# Exibindo as imagens reduzidas
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(imagem_reduzida_50)
plt.title("50%")

plt.subplot(2, 2, 2)
plt.imshow(imagem_reduzida_25)
plt.title("25%")

plt.subplot(2, 2, 3)
plt.imshow(imagem_reduzida_12_5)
plt.title("12.5%")

plt.subplot(2, 2, 4)
plt.imshow(imagem_reduzida_6_25)
plt.title("6.25%")

plt.tight_layout()
plt.show()
```

```python
from PIL import Image

def restaurar_imagem(imagem_reduzida, tamanho_original):
    imagem_restaurada = imagem_reduzida.resize(tamanho_original, Image.BILINEAR)
    return imagem_restaurada

# Carregando as imagens reduzidas
imagem_reduzida_50 = Image.open("/content/imagem_reduzida_50.png")
imagem_reduzida_25 = Image.open("/content/imagem_reduzida_25.png")
imagem_reduzida_12_5 = Image.open("/content/imagem_reduzida_12_5.png")
imagem_reduzida_6_25 = Image.open("/content/imagem_reduzida_6_25.png")

# Obtendo o tamanho original das imagens
tamanho_original = imagem_reduzida_50.size

# Restaurando as imagens reduzidas
imagem_restaurada_50 = restaurar_imagem(imagem_reduzida_50, tamanho_original)
imagem_restaurada_50.save("/content/imagem_reduzida_50.png")

imagem_restaurada_25 = restaurar_imagem(imagem_reduzida_25, tamanho_original)
imagem_restaurada_25.save("/content/imagem_reduzida_25.png")

imagem_restaurada_12_5 = restaurar_imagem(imagem_reduzida_12_5, tamanho_original)
imagem_restaurada_12_5.save("/content/imagem_reduzida_12_5.png")

imagem_restaurada_6_25 = restaurar_imagem(imagem_reduzida_6_25, tamanho_original)
imagem_restaurada_6_25.save("/content/imagem_reduzida_6_25.png")

# Exibindo as imagens restauradas
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(imagem_restaurada_50)
plt.title("Restaurada - 50%")

plt.subplot(2, 2, 2)
plt.imshow(imagem_restaurada_25)
plt.title("Restaurada - 25%")

plt.subplot(2, 2, 3)
plt.imshow(imagem_restaurada_12_5)
plt.title("Restaurada - 12.5%")

plt.subplot(2, 2, 4)
plt.imshow(imagem_restaurada_6_25)
plt.title("Restaurada - 6.25%")

plt.tight_layout()
plt.show()
```
